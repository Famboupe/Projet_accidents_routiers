## Etape 3 : Séparation des datasets



### Import des données

```python
# import des librairies
import pandas as pd
pd.set_option('display.max_columns', None)
pd.options.display.max_info_columns

from sklearn.model_selection import train_test_split
```

```python
# import du dataset
df = pd.read_csv("dataset/dataset_final.csv")
print(df.shape)
df.head()
```

### Séparation du dataset en un train set et un test set

```python
df['index'] = df.index
df_train, df_test = train_test_split(df, test_size=0.2, random_state = 42, stratify=df['grav'])
print("Séparation des labels et targets :", df_train.shape, df_test.shape)
```

### Sauvegarde des datasets

```python
df_train.to_csv('dataset/df_train.csv', index=False)
df_test.to_csv('dataset/df_test.csv', index=False)
```
